```python
import numpy as np
import random as rd
import math
```

```python
n = 1000
A = [[0 for _ in range(n)] for __ in range(n)]
b = [0 for _ in range(n)]
for i in range(n):
  b[i] = rd.uniform(1,10)
  A[i][i] = rd.uniform(1000,1500)
  for j in range(i):
    A[i][j] = rd.uniform(10,15)
    A[j][i] = A[i][j]

# n = 7
# A = [[0.411, 0.421, -0.333, 0.313, -0.141, -0.381, 0.245],
# [0.241, 0.705, 0.139, -0.409, 0.321, 0.0625, 0.101],
# [0.123, -0.239, 0.502, 0.901, 0.243, 0.819, 0.321],
# [0.413, 0.309, 0.801, 0.865, 0.423, 0.118, 0.183],
# [0.241, -0.221, -0.243, 0.134, 1.274, 0.712, 0.423],
# [0.281, 0.525, 0.719, 0.118, -0.974, 0.808, 0.923],
# [0.246, -0.301, 0.231, 0.813, -0.702, 1.223, 1.105]]
# b = [0.096, 1.252, 1.024, 1.023, 1.155, 1.937, 1.673]

# A = [[4, -1, 1],
#      [-1, 4.25, 2.75],
#      [1, 2.75, 3.5]]

# b = [2.6, 1.2, 5.4]

def print_matr(M, r, c):
  for _ in range(r):
    for __ in range(c):
      print(round(M[_][__],2), end = ' ')
    print('\n')

def sum_squares(start, end, iterable):
  s = 0
  for i in range(start, end):
    s += iterable[i]**2
  return s

def sum_prods(start, end, iterable1, iterable2):
  s = 0
  for i in range(start, end):
    s += iterable1[i]*iterable2[i]
  return s

def minor(A,k):
  minor_matrix = [[A[i][j] for j in range(k)] for i in range(k)]
  minor = np.linalg.det(minor_matrix)
  return minor

def positive_definite(A):
  for q in range(1,n+1):
    if (minor(A,q) <= 0):
      return "Matrix is not positive definite"
  return "Ok"

print(positive_definite(A))

#print_matr(A,n,n)
```

```python
def get_L(A):
  for q in range(1,n+1):
    if (minor(A,q) <= 0):
      print("Matrix is not positive definite")
      return 0


  L = [[0 for _ in range(n)] for __ in range(n)]
  L[0][0] = math.sqrt(A[0][0])
  for j in range(1,n):
    L[j][0]=A[j][0]/L[0][0]

  for i in range(1,n-1):
    L[i][i] = math.sqrt(A[i][i] - sum_squares(0,i,L[i]))
    for j in range(i+1, n):
      L[j][i] = (A[j][i] - sum_prods(0,i,L[j], L[i]))/L[i][i]
  L[n-1][n-1] = math.sqrt(A[n-1][n-1]-sum_squares(0,n-1,L[n-1]))
  return L

L = get_L(A)
# print_matr(L,3,3)

def get_solution(L,b):
  y = [0 for _ in range(n)]
  y[0] = b[0]/L[0][0]
  for i in range(1,n):
    y[i] = (b[i] - sum_prods(0,i,L[i],y))/L[i][i]


  x = [0 for _ in range(n)]
  x[n-1] = y[n-1]/L[n-1][n-1]
  for i in range(n-2, -1, -1):
    x[i] = (y[i] - sum_prods(i+1,n,[L[q][i] for q in range(n)], x))/L[i][i]

  return x

if L != 0:
  sol = get_solution(L,b)
  print(f'x = {sol}')
  diff = np.matmul(A,sol) - b
  print(f'Ax - b={diff}')
  print(f'max|Ax-b|={max(abs(diff))}')
```
