```python
import numpy as np
import random as rd
import math
```

```python
n = 100
A = [[0 for _ in range(n)] for __ in range(n)]
b = [0 for _ in range(n)]
for i in range(n):
  b[i] = rd.uniform(1,2)
  for j in range(n):
    A[i][j] = rd.uniform(1,2)

# n = 7
# A = [[0.411, 0.421, -0.333, 0.313, -0.141, -0.381, 0.245],
# [0.241, 0.705, 0.139, -0.409, 0.321, 0.0625, 0.101],
# [0.123, -0.239, 0.502, 0.901, 0.243, 0.819, 0.321],
# [0.413, 0.309, 0.801, 0.865, 0.423, 0.118, 0.183],
# [0.241, -0.221, -0.243, 0.134, 1.274, 0.712, 0.423],
# [0.281, 0.525, 0.719, 0.118, -0.974, 0.808, 0.923],
# [0.246, -0.301, 0.231, 0.813, -0.702, 1.223, 1.105]]
# b = [0.096, 1.252, 1.024, 1.023, 1.155, 1.937, 1.673]



def print_matr(M, r, c):
  for _ in range(r):
    for __ in range(c):
      print(round(M[_][__],2), end = ' ')
    print('\n')

print_matr(A,n,n)

for q in range(1,n+1):
  minor_matrix = [[A[i][j] for j in range(q)] for i in range(q)]
  #print(minor_matrix)
  minor = np.linalg.det(minor_matrix)
  print(minor)
  if minor == 0:
    print('LU decomposition does not exist')
    break
```

```python
def lu_decompose(A):
  L = [[0 for x in range(n)] for y in range(n)]
  U = [[0 for x in range(n)] for y in range(n)]

  for i in range(n):
    for k in range(i, n):              # U
      sum = 0
      for j in range(i):
        sum += (L[i][j] * U[j][k])

      U[i][k] = A[i][k] - sum

    for k in range(i, n):
      if (i == k):
        L[i][i] = 1
      else:
        sum = 0
        for j in range(i):
            sum += (L[k][j] * U[j][i])
        L[k][i] = (A[k][i] - sum) / U[i][i]
  return L,U

L, U = lu_decompose(A)

print(f'L = ')
print_matr(L,n,n)
print('-----------------------------')
print(f'U = ')
print_matr(U,n,n)
print('-----------------------------')

def solve_y(Lw, f):
  y = [0 for _ in range(n)]
  y[0] = b[0]/Lw[0][0]
  for i in range(1,n):
    sum = 0
    for j in range(i):
      sum += (Lw[i][j]*y[j])
    y[i] = (f[i] - sum)

  return y

y = solve_y(L,b)


def solve(Up, Ux):
  x = [0 for _ in range(len(y))]
  x[-1] = Ux[-1]/Up[-1][-1]
  for i in range(len(Ux) - 2, -1, -1):
    sum = 0
    for j in range(i+1,n):
      sum += (Up[i][j]*x[j])
    x[i] = (1/Up[i][i])*(Ux[i] - sum)
  return x


solution = solve(U,y)
print(f'x = {solution}')
print(f'Ax - b = {np.matmul(A,solution) - b} \n max|Ax-b|={max(abs(np.matmul(A,solution) - b))}')
```
