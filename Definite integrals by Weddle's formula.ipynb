```python
a, b = 1.5, 10

def f(x):
  return x**2 - np.log(x) - 4

def analytic(x):
  return 2*(x**2) - sin(x)

```

```python
import math

def Weddle(n):
  h = (b - a) / n
  sum = 0
  for i in range(n+1):
    if i == 0 or i == n:
      sum += 3*h/10*f(a+i*h)
    elif i%6==0:
      sum += 2*3*h/10*f(a+i*h)
    elif i%3 == 0:
      sum += 6*3*h/10*f(a+i*h)
    elif i % 2 == 0:
      sum += 3*h/10*f(a+i*h)
    else:
      sum += 5 * 3 * h/10 * f(a+i*h)
  return sum
epsilon = 10**(-6)
cnt = 1
m = 2
while (abs(Weddle(6*m) - Weddle(6*(m-1))) > epsilon):
  cnt += 1
  m+=1
print(f'iterations: {cnt+1}')
print(f'exact: {analytic(b) - analytic(a)}')
print(f'approximated: {Weddle(6*(m+1))}')
print(abs(Weddle(6*(m)) - Weddle(6*(m-1))))
```
