```python
def eliminate_row(A, elimination_row):
  for k in range(elimination_row):
    num = A[elimination_row, k]
    A[elimination_row] = A[elimination_row] - num*A[k]
    #b[elimination_row] = b[elimination_row] - num*b[k]
  A[elimination_row] = A[elimination_row]/A[elimination_row,elimination_row]
  #b[elimination_row] = b[elimination_row]/A[elimination_row,elimination_row]
  # MISTAKE: ASSIGNING NEW VALUE IN A
  return A, b

def eliminate_upwards_rows(A, eliminated_row):
  for i in range(eliminated_row):
    A[i] = A[i] - A[eliminated_row]*A[i,eliminated_row]
    #b[i] = b[i] - b[eliminated_row]*A[i,eliminated_row]
  return A, b

for i in range(n-1):
  print('-------------------\n')
  print_matr(A,n,n+1)
  # print(f'b = {b}')
  print('\n-------------------')
  A[i] = (A[i]/A[i,i])
  #b[i] = (b[i]/A[i,i])
  #A[i] = new_row
  A, b = eliminate_row(A, i+1)
  A, b = eliminate_upwards_rows(A, i+1)

solution = [A[i,n] for i in range(n)]
print('A = ')
print_matr(A,n,n)

# print('x = ')
# for b_i in solution:
#   print(b_i, '\n')


for_answer = [[0 for i in range(n)] for j in range(n)]

for i in range(n):
  for j in range(n):
    for_answer[i][j] = A_start[i][j]

# diff = np.matmul(for_answer,solution) - b_start
# for c in range(len(diff)):
#   print(f'(Ax)_{c+1} - b_{c+1} = {abs(diff[c,0])} \n')

diff = np.matmul(for_answer,solution) - b_start

print(f'Ax - b = {diff} \n max|Ax-b| = {max(abs(diff))}')
```
