```python
# inputs and utility functions

# n = 20
# A = [[0 for _ in range(n)] for __ in range(n)]
# b = [0 for _ in range(n)]
# for i in range(n):
#   b[i] = rd.randint(-60,60)
#   for j in range(n):
#     A[i][j] = rd.randint(-60,60)

n = 7
A = [[0.411, 0.421, -0.333, 0.313, -0.141, -0.381, 0.245],
[0.241, 0.705, 0.139, -0.409, 0.321, 0.0625, 0.101],
[0.123, -0.239, 0.502, 0.901, 0.243, 0.819, 0.321],
[0.413, 0.309, 0.801, 0.865, 0.423, 0.118, 0.183],
[0.241, -0.221, -0.243, 0.134, 0, 0.712, 0.423],
[0.281, 0.525, 0.719, 0.118, -0.974, 0.808, 0.923],
[0.246, -0.301, 0.231, 0.813, -0.702, 1.223, 1.105]]
b = [0.096, 1.252, 1.024, 1.023, 1.155, 1.937, 1.673]


def print_matr(M, r, c):
  for _ in range(r):
    for __ in range(c):
      print(round(M[_][__],2 ), end = ' ')
    print('\n')

def main_element(M, rows, cols):
  abs_max = 0
  main_row, main_col = 0,0
  for i in range(rows):
    for j in range(cols):
      if abs(M[i][j]) > abs(abs_max):
        abs_max = M[i][j]
        main_row = i
        main_col = j
  return [abs_max, main_row, main_col]

print(main_element(A, n, n), '\n', A)
```

```python
# Solve

resulted_rows = [[0 for _ in range(n+1)] for __ in range(n)]


def processing_straight(M, b, depth):
  t = main_element(M, n - depth, n)
  m, main_row, main_col = t[0], t[1], t[2]
  m_i = [0 for _ in range(n-depth)]
  for i in range(n-depth):
    if i == main_row:
      m_i[i] = 0
    else:
      m_i[i] = -M[i][main_col]/m
  # print(m_i)
  for i in range(n+1):
    if i != n:
      #print(f'{M[main_row][i]}/{m}')
      resulted_rows[depth][i] = M[main_row][i]/m
    else:
      resulted_rows[depth][i] = b[main_row]/m


  # new matrix is built


  b_new = [b[i] + b[main_row]*m_i[i] for i in range(n-depth) if i != main_row]
  #print(b_new)

  M_new = [[M[i][j] + M[main_row][j]*m_i[i] for j in range(n)] for i in range(n-depth) if i != main_row]
  # print_matr(M_new, len(M_new), n)
  # print('------------------------------')
  if depth == n-1:
    return M_new, b
  else:
    return processing_straight(M_new, b_new, depth + 1)

processing_straight(A,b,0)

#print_matr(resulted_rows, n, n+1)


def processing_reverse(M, x, depth):
  excludable_var = 0
  for i in range(n):
    if M[depth][i] == 1.0:
      excludable_var = i
  # print(excludable_var)
  x[excludable_var] = M[depth][-1]

  for i in range(n):
    if i != excludable_var:
      x[excludable_var] -= M[depth][i]*x[i]
  if depth == 0:
    return x
  else:
    #print(x)
    return processing_reverse(M, x, depth-1)

print(f" x = {processing_reverse(resulted_rows, [0 for i in range(n)], n-1)}")
# print(b)
diff = np.matmul(A,processing_reverse(resulted_rows, [0 for i in range(n)], n-1)) - b
for c in range(len(diff)):
  print(f'(Ax)_{c+1} - b_{c+1} = {abs(diff[c])} \n')
```
