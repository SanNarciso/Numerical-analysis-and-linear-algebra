```python
import numpy as np
from sympy import *
from sys import maxsize
```

```python
def f(x):
  return 2 - x*np.exp(x)
x = Symbol('x')
y = 2-x*exp(x)
yprime = y.diff(x)
yprimeprime = yprime.diff(x)

def der(inp):
  f = lambdify(x, yprime)
  return f(inp)

def der_2(inp):
  f = lambdify(x, yprimeprime)
  return f(inp)
```

```python
def get_sign_intervals(function):
    intervals = []
    start = None
    for i in range(len(function)):
        if start is None:
            if function[i] != 0:
                start = i
        elif (function[i] > 0 and function[i-1] <= 0) or (function[i] < 0 and function[i-1] >= 0):
            intervals.append((start, i-1))
            start = i
    if start is not None:
        intervals.append((start, len(function)-1))
    return intervals




st, end = -100, 100
nn = 2000
h = (end - st)/nn
x_values = [round(st + i*h,2) for i in range(nn)]
function_values = [der(x) for x in x_values]

intervals = get_sign_intervals(function_values)
print(x_values[intervals[0][0]], x_values[intervals[0][1]])
print(x_values[intervals[1][0]], x_values[intervals[1][1]])
```
